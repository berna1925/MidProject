Generosity: 관대함 지표

Perceptions of corruption: 부패 인식 지수

Dystopia + residual: 디스토피아 지수 및 잔여값

```python
import pandas as pd

# 파일 경로 설정 (예: 'My Drive/파일이름.xlsx')
file_path = '/content/DataForFigure2.1WHR2023.xls'

# XLS 파일 읽기
df = pd.read_excel(file_path)

# 데이터프레임 출력
df.info()
df.shape
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 'Country name'을 제외하고 이상치를 확인할 열을 선택
columns_to_check = df_cleaned.drop('Country name', axis=1)

# IQR을 이용하여 이상치 계산
Q1 = columns_to_check.quantile(0.25)
Q3 = columns_to_check.quantile(0.75)
IQR = Q3 - Q1

# 이상치로 간주되는 값은 Q1 - 1.5*IQR 이하 또는 Q3 + 1.5*IQR 이상
outliers = ((columns_to_check < (Q1 - 1.5 * IQR)) | (columns_to_check > (Q3 + 1.5 * IQR)))

# 이상치가 있는 데이터를 시각화
plt.figure(figsize=(10, 6))
sns.boxplot(data=columns_to_check)
plt.title('Boxplot for Outlier Detection')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


df_cleaned= df_cleaned.drop('Country name', axis=1)
# "Explained by:" 관련 컬럼 제외
exclude_columns = [
    'Explained by: Log GDP per capita',
    'Explained by: Social support',
    'Explained by: Healthy life expectancy',
    'Explained by: Freedom to make life choices',
    'Explained by: Generosity',
    'Explained by: Perceptions of corruption'
]

df_clean = df_cleaned.drop(columns=exclude_columns)

# 상관관계 행렬 계산
correlation_matrix = df_clean.corr()

# 상관관계 히트맵 시각화
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix of Variables (Excluding 'Explained by:' Columns)")
plt.show()

ladder_correlation = correlation_matrix['Ladder score'].sort_values(ascending=False)

# 결과 출력
print(ladder_correlation)
```
