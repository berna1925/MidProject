```python
import requests

html = requests.get('https://www.numbeo.com/crime/rankings_by_country.jsp?title=2020').text
print(html)
```

```python
from bs4 import BeautifulSoup
from lxml import etree

soup = BeautifulSoup(html, 'html.parser')
content = etree.HTML(str(soup))

security_index = []

for i in range(1, 130):
    numbers = content.xpath(f'//*[@id="t2"]/tbody/tr[{i}]/td[4]')
    if numbers :
        security_index.append(numbers[0].text)

# security_index
```

```python
country_name = []

for i in range(1, 130):
    names = content.xpath(f'//*[@id="t2"]/tbody/tr[{i}]/td[2]')
    if names:
        country_name.append(names[0].text.strip())

# country_name
```

```python
import pandas as pd

df_2020 = pd.DataFrame({'country name' : country_name,
                       'security index' : security_index})

df_2020.to_excel('d_2020.xlsx', index=False)
```

# 상관관계 분석(히트맵)

```python
import pandas as pd

df = pd.read_excel('피벗 테이블 제작용 파일.xls')
df
```

```python

df_c = df.copy()

columns = ['Ladder score',
           'Explained by: Log GDP per capita',
           'Explained by: Social support',
           'Explained by: Healthy life expectancy',
           'Explained by: Freedom to make life choices',
           'Explained by: Generosity',
           'Explained by: Perceptions of corruption',
           'security rate']

df_c_ = df_c[columns]

corr = df_c_.corr()

import seaborn as sns

sns.heatmap(corr, fmt='.2f', annot=True)
```

```python
import pandas as pd

df = pd.read_excel('연도별 이민자 수와 안전 지수.xlsx')
columns = ['Immigrant application', 'Security index', 'Ladder score']

df_k = df.copy()
df_k = df_k[columns]

import seaborn as sns
sns.heatmap(df_k.corr(), fmt='.2f', annot=True)

# 이민자 수의 증가가 사회 안전 지수를 저해한다는 근거가 없음
# 태블로 확인 결과 전 세계적으로 보면 남미, 프랑스에서 안전 지수와 행복 지수가 반대 방향으로 변화한 것이 확인
# 이민자 수의 증가가 사회 안전 지수를 저해해 행복 지수를 저해할 수 있다는 소가설은 근거가 부족

# 논문 등 자료 검색 결과 중남미에서는 높은 사회적 연대감이 안전 지수와 행복 지수 사이 양의 상관관계가 반박되는 예시로 확인
```
