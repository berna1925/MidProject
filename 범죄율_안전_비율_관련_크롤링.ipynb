```python
import requests

html = requests.get('https://www.numbeo.com/crime/rankings_by_country.jsp?title=2020').text
print(html)
```

```python
from bs4 import BeautifulSoup
from lxml import etree

soup = BeautifulSoup(html, 'html.parser')
content = etree.HTML(str(soup))

security_index = []

for i in range(1, 130):
    numbers = content.xpath(f'//*[@id="t2"]/tbody/tr[{i}]/td[4]')
    if numbers :
        security_index.append(numbers[0].text)

# security_index
```

```python
country_name = []

for i in range(1, 130):
    names = content.xpath(f'//*[@id="t2"]/tbody/tr[{i}]/td[2]')
    if names:
        country_name.append(names[0].text.strip())

# country_name
```

```python
import pandas as pd

df_2020 = pd.DataFrame({'country name' : country_name,
                       'security index' : security_index})

df_2020.to_excel('d_2020.xlsx', index=False)
```
